# challenge 8
Modify the "rating" column to group similar ratings together. For example, group all "PG-13", "TV-14", and "TV-PG" ratings under a new label "Teen". and do a pivot table again

```python
rating_mapping = {
    'PG-13': 'Teen',
    'TV-14': 'Teen',
    'TV-PG': 'Teen',
    'R': 'Adult',
    'TV-MA': 'Adult',
    'G': 'Kids',
    'TV-Y': 'Kids',
    'TV-Y7': 'Kids',
    'PG': 'Family'
}

# Modify the 'rating' column to group similar ratings together
df['rating_grouped'] = df['rating'].replace(rating_mapping)

# Create a pivot table summarizing the distribution of grouped ratings by year and type
pivot_table = pd.pivot_table(df, 
                             values='show_id', 
                             index='release_year', 
                             columns=['type', 'rating_grouped'], 
                             aggfunc='count', 
                             fill_value=0)

# Display the pivot table
print(pivot_table)
```
